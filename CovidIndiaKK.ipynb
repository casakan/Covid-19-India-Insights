```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
print("Modules are loaded")
```

```python
path = "C:/Users/casak/KKlab/Covid/data/datasets_557629_1255329_StatewiseTestingDetails.csv"
corona_india_df = pd.read_csv(path)
corona_india_df.head()
```

```python
corona_india_df.shape
```

```python
corona_india_aggregated = corona_india_df.groupby(["State"]).sum()
corona_india_aggregated
```

```python
corona_india_aggregated["PercentagePositive"] = (corona_india_aggregated["Positive"]/corona_india_aggregated["TotalSamples"])*100
corona_india_aggregated.round(2)
```

```python
corona_india_aggregated = corona_india_aggregated.sort_values(["TotalSamples"], ascending = False)
corona_india_aggregated
```
